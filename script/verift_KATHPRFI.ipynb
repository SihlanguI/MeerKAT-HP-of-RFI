```python
import katdal
import numpy as np
import pandas as pd
import kathprfi_single_file as kathp

csv_file = pd.read_csv('sci_Imaging_L_2024-01-01T00:00:00Z_2024-01-31T00:00:00Z.csv')
filename = csv_file['FullLink']
vis = [katdal.open(obs) for obs in filename]



```

```python
import time as tme
import xarray as xr
import zarr
import logging
import six
import ast
import numba
from numba import prange
import os




good_tags = set(["target", "bpcal", "delaycal","fluxcal","gaincal","polcal"])
good_targets = []
count = 0
good_targets = []

def initialize_logs():
    """
    Initialize the log settings
    """
    logging.basicConfig(format='%(message)s', level=logging.INFO)

for v, name in zip(vis, filename):
    #average_factor = v.shape[1] // 4096
    # if average_factor > 1:
    #     v_reshape = v.reshape(v.shape[0], average_factor, 4096, v.shape[2] )
    #     v = v_reshape.mean(axis=1)

    # Initializing 5-D arrays
    master = np.zeros((24, 4096, 2016, 8, 24), dtype=np.uint16)
    counter = np.zeros((24, 4096, 2016, 8, 24), dtype=np.uint16)
    s = tme.time()
    clean_ants = kathp.remove_bad_ants(v)

    logging.info('Adding file {} : {}'.format(count, name))
    if v.shape[1] == 4096:
        for tar in v.target_indices:
            target = v.catalogue.targets[tar]
            if len(good_tags.intersection(target.tags)) > 1:
                good_targets.append(target)
                v.select(corrprods='cross', pol='HH', scans='track', ants=clean_ants,
                flags='cal_rfi', targets = good_targets)
                good_flags = v.flags
                if good_flags.shape[0] * good_flags.shape[1] * good_flags.shape[2] != 0:
                    # Updating the array
                    ntime = v.shape[0]
                    time_step = 1
                    if ntime <= time_step:
                        time_step = ntime
                    nant = 64
                    Bl_idx = kathp.get_bl_idx(v, nant)
                    elbins = np.linspace(10, 80, 8)
                    azbins = np.arange(0, 360, 15)
                    el, az = kathp.get_az_and_el(v)
                    logging.info('Start to update the master and counter array')
                    for tm in six.moves.range(0, ntime, time_step):
                        time_slice = slice(tm, tm + time_step)
                        flag_chunk = good_flags[time_slice].astype(int)
                        # average flags from 32k to 4k mode.
                        if good_flags.shape[1] == 32768:
                            flag_chuck = kathp.NewFlagChunk(flag_chunk)
                        Time_idx = kathp.get_time_idx(v)[time_slice]
                        El_idx = kathp.get_el_idx(el, elbins)[time_slice]
                        Az_idx = kathp.get_az_idx(az, azbins)[time_slice]
                        master, counter = kathp.update_arrays(Time_idx, Bl_idx, El_idx, Az_idx,
                                                                flag_chunk, master, counter)
                    logging.info('{} s has been taken to update file number {}'.format(count,
                                                                                        tme.time()
                                                                                        - s))
                    logging.info('Creating Xarray Dataset')
                    ds = xr.Dataset({'master': (('time', 'frequency', 'baseline', 'elevation',
                                                    'azimuth'), master),
                    'counter': (('time', 'frequency', 'baseline', 'elevation', 'azimuth'), counter)},
                    {'time': np.arange(24), 'frequency': v.freqs, 'baseline': np.arange(2016),
                        'elevation': np.linspace(10, 80, 8), 'azimuth': np.arange(0, 360, 15)})
                    logging.info('Saving dataset')
                    ds.to_zarr(str(name[46:56]))
                    logging.info('Dataset has been saved')
                    logging.info('File has been saved')

                    count+=1

                # print(f"{name[46:56]} these are good targets: {target} nchan {v.shape[1]}")
                # count+=1
    
count
```
